```python
import pandas as pd
```

```python
aol = pd.read_csv("data/AOL/user-ct-test-collection-02.txt", names=["uid","query", "time", "rank", "url"], skiprows=[0], sep="\t", error_bad_lines=False)
```

```python
aol['time']=pd.to_datetime(aol['time'])
```

```python
aol = aol.drop_duplicates(['uid', 'query'], keep='first')
```

```python
aol = aol[~aol["query"].str.contains(".com", na=False)]
aol = aol[~aol["query"].str.contains("www", na=False)]
```

```python
aol.index = aol['time']
```

```python
for name, group in aol.groupby(by=[aol.uid, aol.index.year, aol.index.month, aol.index.day, aol.index.hour]):
    if len(group) > 10:
        print(group['query'])
        print()
```
